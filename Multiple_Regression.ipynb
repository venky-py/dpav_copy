```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

 Separate the input and output variables

```python
# input -> disp, hp, wt
x= df[['disp','hp','wt']]
# output -> mpg
y=df['mpg']
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
x.corrwith(y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
disp= 221
hp= 102
wt= 2.91
```

```python
reg.predict([[221, 102, 2.91]])
```

```python
y= np.sum(reg.coef_*[disp,hp,wt]) + reg.intercept_
```

```python
y
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.xlabel('disp')
plt.ylabel('mpg')
plt.scatter(df['disp'], df['mpg'], color='r')
plt.subplot(2,2,2)
plt.xlabel('hp')
plt.ylabel('mpg')
plt.scatter(df['hp'], df['mpg'], color='g', marker='D')
plt.subplot(2,2,3)
plt.xlabel('wt')
plt.ylabel('mpg')
plt.scatter(df['wt'], df['mpg'], color='b', marker='^');
```
