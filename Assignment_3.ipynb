```python
cricket['Avg'] = cricket.runs/cricket.matches
```

```python
cricket
```

```python
new_cricket = pd.DataFrame({
    'name':["Anil Dalpat", "Rohan Kanhay", "Avdhoot Dighe","Bahubali", "Leeladhar", "Pradyumna", "Dinesh Roy",
            "Parmeshwar", "Ali Durrani", "Litesh Singh"],
    'Zone':['N','S','W','N','W','NE','E','S','W','W']
})   
```

```python
new_cricket
```

```python
merged = cricket.merge(new_cricket, on='name')
```

```python
merged
```

```python
gr = merged.groupby('Zone')
```

```python
gr.groups
```

```python
gr.get_group('W')
```

```python
gr.value_counts()
```

```python
gr.describe()
```

```python
merged.to_csv('cricket_merged', index=False)
```
