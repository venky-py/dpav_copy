Bayes theorem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
# Separate the features and labels
x= df[['Age','EstimatedSalary']]
y=df['Purchased']
```

```python
# import the class
from sklearn.naive_bayes import GaussianNB
```

```python
# create the object
nb= GaussianNB()
```

```python
# Apply the data / train the algorithm
nb.fit(x,y)
```

```python
new = pd.DataFrame({
    'Age':[46,23,52],
    'EstimatedSalary':[67000,38700,140000]
})
```

```python
new
```

```python
nb.predict_proba(new)
```

```python
probs=nb.predict_proba(new)
```

```python
probs.argmax(axis=1)
```

```python
new1 = pd.DataFrame({
    'Age':[46,23,52,21,40,33],
    'EstimatedSalary':[67000,38700,140000,50000,370000,20000]
})
```
