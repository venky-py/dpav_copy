```python
x=df.drop('mpg',axis=1)
```

```python
correlation_matrix = x.corr()
correlated_features = set()
```

```python
correlation_matrix
```

```python
%precision 2
for i in range (len(correlation_matrix)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i,j])>0.8:
            correlated_features.add(correlation_matrix.columns[i])
```

```python
correlated_features
```

```python
df.drop(correlated_features, axis = 1, inplace=True)
```

```python
df
```

```python
df.corr()
```

Rank Correlation

```python
from scipy.stats import rankdata
```

```python
x=np.array([4,11,15,21,28,34,39,35,47,55])
y=np.array([10,22,31,36,48,51,58,61,69,77])
z=np.array([92,83,89,56,61,56,48,41,26,19])
p=np.array([56,44,27,65,38,61,23,75,36,58])
```

```python
x
```

```python
rankdata(x)
```

```python
rankdata(y)
```

```python
rankdata(p)
```

```python
rankdata(z)
```

Shearman's Correlation

```python
from scipy.stats import spearmanr,pearsonr
```

```python
df= pd.DataFrame({
    'X':x,
    'Y':y,
    'Z':z,
    'P':p
})
```

```python
spearmanr(x,y)
```

```python
pearsonr(rankdata(x), rankdata(y))
```

```python
df.corr()
```

```python
df.corr(method='spearman')
```

```python
a= np.array([12,15,27,48,56,79])
b= np.array([10,23,33,37,56,89])
```

```python
spearmanr(a, b)
```

```python
pearsonr(a, b)
```

Kendall's Tau

```python
from scipy.stats import kendalltau
```

```python
kendalltau(x,y)
```

```python
kendalltau(x,z)
```

```python
kendalltau(x,p)
```

```python
df.corr(method='kendall')
```
