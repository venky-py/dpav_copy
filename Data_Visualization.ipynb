Bar Plot

```python
enrol= pd.DataFrame({
    'Students': [67,89,106,179,47,122,100],
    'Languages': ['C','C++','Java','Python','R','PHP','JavaScript']
})
```

```python
enrol
```

```python
plt.title('Student Enrollment for Training')
plt.xlabel('Language')
plt.ylabel('Total Students')
plt.bar(enrol['Languages'],enrol['Students'], color= 'magenta')
```

```python
plt.title('Student Enrollment for Training')
plt.xlabel('Language')
plt.ylabel('Total Students')
plt.bar(enrol['Languages'],enrol['Students'], color= 'cyan')
plt.plot(enrol['Languages'], enrol['Students'], marker = '*')
```

```python
plt.title('Student Enrollment for Training')
plt.ylabel('Total Students')
plt.xlabel('Language')
plt.barh(enrol['Students'],enrol['Languages'], color= 'cyan')
```

```python
enrol= pd.DataFrame({
    'Students': [67,89,106,179,47,122,100],
    'Languages': ['C','C++','Java','Python','R','PHP','JavaScript'],
    'AvgMarks': [67.45,69.23,72.45,78.23,63.27,69.34,74.23]
})
```

```python
enrol
```

```python
plt.title('Student Enrollment for Training')
plt.xlabel('Language')
plt.ylabel('Total Students')
plt.bar(enrol['Languages'],enrol['Students'], color= 'navy', align='edge', width=0.5, label='Total Student')
plt.bar(enrol['Languages'],enrol['AvgMarks'], color= 'skyblue', align='center', width=0.25, label='Avgmarks')
plt.legend()
```

Pie Chart

```python
plt.figure(figsize=(6,6))
plt.title('% Share of students')
plt.pie(enrol['Students'], labels=enrol['Languages'], autopct='%2.2f%%');
```

```python
plt.figure(figsize=(6,6))
plt.pie(sna['Purchased'].value_counts(), labels=['Purchased', 'Not Purchased'], autopct='%.2f%%');
```

Donut Chart

```python
my_circle = plt.Circle((0,0), 0.7, color='white')

plt.pie(enrol['Students'], labels=enrol['Languages'], autopct='%.2f%%');
p = plt.gcf()
p.gca().add_artist(my_circle)
```
